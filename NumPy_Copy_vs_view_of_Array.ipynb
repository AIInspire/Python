# NumPy Copy and view of Array

The NumPy array is a data structure consisting of two parts:

1.   the contiguous data buffer with the actual data elements
2.   the metadata that contains information about the data buffer(data type, strides, and other important information)

image.png

**1. View:**


*   a view of the original ndarray and the view does not own the data.
*   view() function
*   Changes are made to one of them affects the other




```python
import numpy as np

array = np.arange(6)
print("Original Array: ", array)

viewed_array = array.view()
print("View: ", viewed_array)

array[0] = 10

print("\nOriginal Array: ", array)
print("View: ", viewed_array)
```

```python
import numpy as np

array = np.arange(6)
print("Original Array: ", array)

viewed_array = array[0:6]
print("View: ", viewed_array)

array[0:2] = [16, 17]

print("\nOriginal Array: ", array)
print("View: ", viewed_array)
```

**2. Copy**


*   completely a new array and the copy owns the data.
*   copy() function
*   Changes are made to one of them does not affect the other





```python
import numpy as np

array = np.arange(6)
print("Original Array: ", array)

coped_array = array.copy()
print("Copy: ", coped_array)

array[0] = 10

print("\nOriginal Array: ", array)
print("Copy: ", coped_array)
```

```python
import numpy as np
x = np.arange(9).reshape(3, 3)
print(x)

y = x[[1, 2]]
print(y)

y.base is None
```

```python
x[[1, 2]] = [[10, 11, 12], [13, 14, 15]]

print(x)
print("\n")
print(y)
```

```python
import numpy as np
x = np.arange(9)
print(x)
print("\n")

y = x.reshape(3, 3)
print(y)
print("\n")

y.base  # .reshape() creates a view
```

```python
z = y[[2, 1]]
print(z)

z.base is None  # advanced indexing creates a copy
```

```python
z.base
```

**View**

* **No data duplication**: The original data buffer remains unchanged.

* **Shared data**: Any modifications made to the view will be reflected in the original array.

* **Fast**: Creating a view is quick because it only involves changing metadata.




**Copy**

* **Data duplication**: Both the data buffer and metadata are duplicated.

* **Independent data**: Modifications to the copy will not affect the original array.

* **Memory consumption**: Creating a copy requires more memory and takes more time.
